##1. Carregando o Dataset:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando o dataset
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQ-HGxRF164bbYUgeX5LC_0ac9ekeWqC2-xFkTiX1-ki8EDV-w5dx56by3GgMMcAWyN6piXHX0OiK3d/pub?gid=1848033570&single=true&output=csv"
df = pd.read_csv(url) #DataFrame
display(df)
```

## 2 - Transformação dos Dados da coluna demissão:

```python
#Transformando demissão em binario
df["Demissão"] = df["Demissão"].map({"Sim": 1, "Não": 0})
display(df)
```

## 3 - Criação de Coluna Salário Mensal:

```python
# Criando a coluna df['salarioMensal'] = df['Salário por hora'] * df['Jornada padrão de trabalho']
df['salarioMensal'] = df['Salário por hora'] * df['Jornada padrão de trabalho']
display(df)
```

## 4 - Quantidade de pedido de demissão

---



```python
# Criando o countplot com a correção
plt.figure(figsize=(8, 6))
sns.countplot(x=df["Demissão"], hue=df["Demissão"], palette=["red", "blue"], legend=False)
plt.title("Contagem de Demissões")
plt.xlabel("Demissão")
plt.ylabel("Contagem")
plt.show()
```

```python
# Selecionando apenas as colunas quantitativas (numéricas)
quantitative_columns = df.select_dtypes(include=['int64', 'float64'])

# Exibindo as colunas quantitativas
print("Colunas quantitativas:")
display(quantitative_columns)
```

```python
# Contando os valores da coluna 'Demissão'
contagem = df['Demissão'].value_counts()
# Calculando as porcentagens
porcentagem = df['Demissão'].value_counts(normalize=True) * 100
# Exibindo os resultados usando iloc para evitar o warning
print(f"Funcionários que ficaram (Não): {contagem.iloc[0]} ({porcentagem.iloc[0]:.2f}%)")
print(f"Funcionários que pediram demissão (Sim): {contagem.iloc[1]} ({porcentagem.iloc[1]:.2f}%)")
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Demissão', y='Distância de Casa', palette='coolwarm')
plt.title('Distribuição da Distância de Casa por Demissão')
plt.xlabel('Demissão (0: Não, 1: Sim)')
plt.ylabel('Distância de Casa (km)')
plt.show()
```

```python
# Contar a frequência de cada valor da coluna 'Equilíbrio vida trabalho'
equilibrio_counts = df['Equilíbrio vida trabalho'].value_counts()

# Plotando o gráfico de pizza
plt.figure(figsize=(8,6))
plt.pie(equilibrio_counts, labels=equilibrio_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99', '#ff6666'])
plt.title('Distribuição do Equilíbrio Vida-Trabalho')
plt.axis('equal')  # Garantir que o gráfico seja um círculo
plt.show()
```

```python
# Calcular a mediana da idade para funcionários que pediram e não pediram demissão
mediana_idade_demitidos = df[df["Demissão"] == 1]["Idade"].median()
mediana_idade_nao_demitidos = df[df["Demissão"] == 0]["Idade"].median()

# Calcular a media da idade para funcionários que pediram e não pediram demissão
media_idade_demitidos = df[df["Demissão"] == 1]["Idade"].mean()
media_idade_nao_demitidos = df[df["Demissão"] == 0]["Idade"].mean()

print(f'Mediana Idades: Demitidos {mediana_idade_demitidos}, Não Demitimos: {mediana_idade_nao_demitidos}')
print(f'Media Idades: Demitidos {media_idade_demitidos:.2f}, Não Demitimos: {media_idade_nao_demitidos:.2f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criar boxplot para comparar distribuições
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Demissão"], y=df["Idade"], palette=["red", "blue"])
plt.xticks([0, 1], ["Não demitidos", "Demitidos"])
plt.xlabel("Status de Demissão")
plt.ylabel("Idade")
plt.title("Distribuição da Idade entre Funcionários Demitidos e Não Demitidos")
plt.show()
```
